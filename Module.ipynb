```python
import datetime

x = datetime.datetime.now()

print(x.year)           #strftime string format of time
print(x.strftime("%A")) # %A is a placeholder that represents the full name of the weekday
```

```python
import datetime
x = datetime.datetime.now()
print("short form weekday:",x.strftime("%a"))
print("Full form weekday: ",x.strftime("%A"))
print("digit form weekday: ",x.strftime("%w"))
print("day in month: ",x.strftime("%d"))
print("day in month in short: ",x.strftime("%b"))
print("day in month in full: ",x.strftime("%B"))
print("digit form of month: ",x.strftime("%m"))
print("year in short: ",x.strftime("%y"))
print("year in full: ",x.strftime("%Y"))
```
